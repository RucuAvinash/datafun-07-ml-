```python
# Cell 1: Imports
import random
import matplotlib.pyplot as plt
import pandas as pd
from deck import DeckOfCards 
from card import Card 

# Set display options for cleaner output
# pd.set_option('display.max_rows', None) # Uncomment if you want to see all rows
pd.set_option('display.float_format', lambda x: '%.2f' % x)
print("Setup complete. Classes and libraries imported.")
```

```python
# Cell 2: Basic Deck Test
# Initialize and display the unshuffled deck
deck = DeckOfCards()
print("--- Unshuffled Deck (First 8 Cards) ---")
print('\n'.join(str(c) for c in deck._deck[:8]))

# Shuffle the deck
deck.shuffle()
print("\n--- Shuffled Deck (First 8 Cards) ---")
print('\n'.join(str(c) for c in deck._deck[:8]))
```

```python
# Cell 3: Simulation
NUM_SIMULATIONS = 1000  # Number of times to draw a full deck of 52 cards

# Lists to store results
faces_dealt = []
suits_dealt = []

# Perform the simulation
for _ in range(NUM_SIMULATIONS):
    deck.shuffle()
    # Deal all 52 cards
    for _ in range(DeckOfCards.NUMBER_OF_CARDS):
        card = deck.deal_card()
        if card:
            faces_dealt.append(card.face)
            suits_dealt.append(card.suit)
```

```python
# Cell 4: Card Face Distribution
# Convert to a Pandas Series and count frequencies
face_counts = pd.Series(faces_dealt).value_counts().reindex(Card.FACES)

plt.figure(figsize=(12, 6))
face_counts.plot(kind='bar', color='skyblue')

# Calculate the expected mean count
expected_mean = NUM_SIMULATIONS * 4 
plt.axhline(expected_mean, color='r', linestyle='--', label=f'Expected Mean ({expected_mean:.0f} per face)')

plt.title(f'Distribution of Card Faces Over {NUM_SIMULATIONS} Full Deck Deals')
plt.xlabel('Card Face')
plt.ylabel('Total Count')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# Display the data
print("Face Counts Data:")
print(face_counts)
```
